```python
# 2. Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.
def arithmetic_progression_matrix(N, first_element, difference):
    progression = np.array([first_element + i * difference for i in range(N)])
    return np.diag(progression)

N = 10
first_element = 2.0
difference = 0.9

print(arithmetic_progression_matrix(N, first_element, difference))
```

```python
# 5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.
chess = np.zeros((5, 5), dtype=int)
chess[1::2, 1::2] = 5
chess[::2, ::2] = 5

triangle = np.triu(np.ones((5, 5), dtype=int) * 5)

det_chess = np.linalg.det(chess)
det_triangle = np.linalg.det(triangle)

inv_chess, inv_triangle = None, None

# Если детерминант не равен нулю, то можно найти обратную матрицу
if det_chess != 0:
    inv_chess = np.linalg.inv(chess)

if det_triangle != 0:
    inv_triangle = np.linalg.inv(triangle)

print(f"Матрица 5 в шахматном порядке:\n{chess}")
print(f"Детерминант шахмат: {det_chess}")
print(f"Обратная матрица шахматы:\n{inv_chess}" if inv_chess is not None else "Обратной матрицы для шахмат нет")

print(f"\nВерхнетреугольная:\n{triangle}")
print(f"Детерминант верхнетреугольной:\n{det_triangle}")
print(f"Обратная матрица верхнетреугольная:\n{inv_triangle}" if inv_triangle is not None else "Обратной матрицы нет")
```

```python
# 9. Удалите из таблицы столбцы, содержащие идентификаторы, переименуйте все оставшиеся названия колонок на русском языке.
delete_titles = ['Id', 'DistrictId']

for title in delete_titles:
    if title in data.columns:
        data = data.drop(columns=[title])

english_to_russian = {
    'Rooms':         'Комнаты',
    'Square':        'Площадь',
    'LifeSquare':    'Жилая площадь',
    'KitchenSquare': 'Кухонная площадь',
    'Floor':         'Этаж',
    'HouseFloor':    'Этажи',
    'HouseYear':     'Год постройки',
    'Ecology_1':     'Экология_1',
    'Ecology_2':     'Экология_2',
    'Ecology_3':     'Экология_3',
    'Social_1':      'Социальное_1',
    'Social_2':      'Социальное_2',
    'Social_3':      'Социальное_3',
    'Healthcare_1':  'Здравоохранение_1',
    'Helthcare_2':   'Здравоохранение_2',
    'Shops_1':       'Магазины_1',
    'Shops_2':       'Магазины_2',
    'Price':         'Цена',
}

data.columns = list(map(lambda title: title if title not in english_to_russian.keys()
                   else english_to_russian.get(title),
                   data.columns))

print(data.head())
```

```python
# 11. Выделите в отдельную таблицу последние десять строк. Уберите в ней столбец с ценой. Склейте ее с первоначальной таблицей при помощи concat. Заполните отсутствующие значения цены средним по таблице.

# Последних 10 строк
last_ten_rows = data.tail(10)
# Удаление столбца с ценой
last_ten_rows = last_ten_rows.drop(columns=['Цена'])
# Склейка с таблицей
combined_data = pd.concat([data, last_ten_rows], ignore_index=True)
# Заполнение отсутствующих значений цены средним значением
target_column = 'Цена'
# combined_data[target_column].fillna(data[target_column].mean(), inplace=True)
combined_data.fillna({target_column: data[target_column].mean()}, inplace=True)
# Проверка
print(combined_data.tail(15))
```

```python
# 12. Выделите пять последних колонок в отдельную таблицу. Удалите в ней строки, в которых цена ниже среднего. Присоедините эту таблицу к изначальной (выберите самый подходящий тип соединения).

# Выделение 5 последних колонок
last_five_columns = data.iloc[:, -5:]
# Удаление строк
target_column = 'Цена'
filtered_last_five = last_five_columns[data[target_column] >= data[target_column].mean()]
# Присоединение таблицы к изначальной (используем метод объединения 'inner')
combined_data = pd.merge(data, filtered_last_five, left_index=True, right_index=True, how='inner')
# Проверка результата
print(combined_data.tail(10))
```

```python
# 13. Выведите таблицу, содержащую среднюю цену и количество квартир на каждом этаже из первоначального набора данных.

# Группировка данных по этажам
floors_statistics = data.groupby('Этаж').agg(Средняя_цена=('Цена', 'mean'), Количество_квартир=('Цена', 'count')).reset_index()
print(floors_statistics)
```

```python
# 17. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

path_to_csv = '/home/lagovsp/ml-tasks/ML0.1 intro/data.csv'
data = pd.read_csv(path_to_csv)

# Подсчёт количества квартир в зависимости от количества комнат
room_counts = data['Rooms'].value_counts()
print(room_counts)

# Поиск сектора с наибольшим числом квартир
max_count_index = room_counts.idxmax()

# Построение круговой диаграммы
plt.figure(figsize=(5, 5))
wedges, texts, autotexts = plt.pie(
    room_counts,
    labels=list(map(int, room_counts.index)),
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.2 if room == max_count_index else 0 for room in room_counts.index],
    colors=sns.color_palette('deep')
)

# Настройка
plt.setp(autotexts, size=12, weight="bold", color="white")
plt.setp(texts, size=20)
plt.title('Количество квартир в процентах в зависимости от количества комнат', fontsize=16, weight="bold")
plt.show()
```

```python
# 22. При помощи сетки графиков PairGrid визуализируйте попарные отношения признаков Rooms, Square, HouseFloor, HouseYear, Price следующим образом: на диагонали - гистограммы, под диагональю - ядерные оценки плотности, над диагональю - диаграммы рассеяния. По результатам визуализации сделайте выводы.

# Определение признаков для визуализации
features = ['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']

# Создание сетки графиков PairGrid
g = sns.PairGrid(data[features], diag_sharey=False)

# Диагональ: гистограммы
g.map_diag(sns.histplot, kde=True, color='skyblue', bins=2)

# Под диагональю: ядерные оценки плотности
g.map_lower(sns.kdeplot, cmap="Blues", fill=True, alpha=0.5)

# Над диагональю: диаграммы рассеяния
g.map_upper(sns.scatterplot, color='orange', alpha=0.6)

# Настройка заголовков
g.fig.suptitle('Парные отношения признаков', y=1.02, fontsize=16, weight='bold')
plt.subplots_adjust(top=0.9)  # Настройка для отступа заголовка

# Отображение графика
plt.show()
```
